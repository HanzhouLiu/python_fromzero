```python
def debug(n, plans):
  streets = set()
  for plan in plans:
    for street in range(plan[0], plan[1] + 1):
      if street not in streets:
        streets.add(street)
  return n - len(streets)
```

```python
import re
def path(home, office, edges):
  myhash = {}
  for edge in edges:
    if edge[0] not in myhash:
      myhash[edge[0]] = [[edge[1], edge[2]]]
    else:
      myhash[edge[0]].append([edge[1], edge[2]])
  costs = []
  def dfs(node, curCost):
    if node == office:
      costs.append(curCost)
      return 
    for nextNode, cost in myhash[node]:
      dfs(nextNode, curCost + cost)
      return
  dfs(home, 0)
  print(costs)
```
